# Projekt-Download

Dieses Notebook erstellt eine ZIP-Datei Ihres gesamten Projektverzeichnisses,
die Sie herunterladen und als Abgabe einreichen können.

**Führen Sie einfach die Zelle unten aus – fertig.**


```python
import shutil, os, sys

PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.insert(0, PROJECT_ROOT)
from config import GROUP_ID

src     = "/home/jovyan/mlops-gasprices"
dst     = f"/home/jovyan/{GROUP_ID}-abgabe"

shutil.make_archive(dst, "zip", src)

zip_path = dst + ".zip"
size_mb  = os.path.getsize(zip_path) / 1024 / 1024
print(f"ZIP erstellt : {zip_path}")
print(f"Größe        : {size_mb:.1f} MB")
print()
print("Download-Anleitung:")
print("  1. Linke Sidebar → Ordner-Symbol (Datei-Browser)")
print("  2. Nach /home/jovyan/ navigieren (Pfeil nach oben, falls nötig)")
print(f"  3. Rechtsklick auf '{GROUP_ID}-abgabe.zip' → Download")
```
